# Part-I A

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#open the csv file
df=pd.read_csv("ab40c054-5031-4376-b52e-9813e776f65e.csv")
df
```

```python
#Let's check the Dimenssions of the Dataframe
df.shape
```

```python
#Remove the rows: '(% Growth over the previous year)' and 'GSDP - CURRENT PRICES (` in Crore)' for the year 2016-17.

df1=df[df.Duration !='2016-17']
df1
```

```python
# taking only the required years(2013-14,2014-15,2015-16)
df2=df1.iloc[6:9]
df2
```

```python

#Lets's Calculate the mean

total_GDP=df.iloc[[7,8,9]].mean().sort_values()

#Drop the column where NAN
total_GDP=total_GDP.drop(labels=['West Bengal1'])

total_GDP
```

```python
#plotting the graph
fig=plt.figure(figsize=(12,8))
ax=total_GDP.plot(kind='barh')
plt.show()

```

```python
#Which states have been growing consistently fast, and which ones have been struggling?

#for analysing the above, we have taken the top five and bottom five states on basis of the average mean for three years.
# and  we can see that Mizoram and tripura are fast growing and Goa and Meghalya are struggling

df3=total_GDP[0:5]
df4=total_GDP[-5:]

plt.figure(figsize=(14,8))
           
plt.subplot(121)
plt.title("Struggling")
plt.plot(df3,color='blue', marker='o', linestyle='dotted',linewidth=3, markersize=8)
plt.xlabel("STATE")
plt.ylabel("GDP")

plt.subplot(122)
plt.title("Fast Growing")
plt.plot(df4, color='blue', marker='o', linestyle='dotted',linewidth=3, markersize=8)
plt.xlabel("STATE")
plt.ylabel("GDP")

plt.show()
```

```python
# take the 2015-16 year data
df5=df1.iloc[4:5]
#sorted the values
df6=pd.DataFrame(df5.mean(axis=0).sort_values ())

#remove NAN values
df7=df6.iloc[0:23]



plt.figure(figsize=(14,8))

plt.subplot(121)
plt.title("Top 5")
plt.plot(df7.tail(),color='red', marker='o', linestyle='dotted',linewidth=3, markersize=8)
plt.xlabel("STATE")
plt.ylabel("GDP")

plt.subplot(122)
plt.title("Bottom 5")
plt.plot(df7.head(), color='red', marker='o', linestyle='dotted',linewidth=3, markersize=8)
plt.xlabel("STATE")
plt.ylabel("GDP")

plt.show()
```

```python
# import libraries
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from glob import glob
#it grab all the csv files from the directory 
list_of_files=glob("*.csv")
l=len(list_of_files)
list_of_files=list_of_files[1:l-1]
df_list=[]
for f in list_of_files:
  
    df_1=pd.read_csv(f,index_col=None , header=0, encoding="ISO-8859-1")
    df_1["State"]=f.split("-")[1]
    df_list.append(df_1)
full_df=pd.concat(df_list,axis=0, ignore_index=True,sort=False)




    
    
    
   

```

```python
#create pivot table only for 2014-15 duration
pivot=full_df.pivot_table(index="Item" ,columns=["State"], values=["2014-15"])
pivot
```

```python
#Filter out  the union territories and perform analysis by the center
df_2=pivot.iloc[[9,15,17,24,31],:]
df_3=df_2.transpose()
df_3
```

```python
#top 5
df_4=df_3.sort_values("Per Capita GSDP (Rs.)", ascending=False)
new_df=df_4.head()

plt.figure(figsize=(9,4))
sns.set_style("whitegrid")
sns.barplot(x=new_df.index,y=new_df["Per Capita GSDP (Rs.)"],data=new_df)
plt.ylabel("Per Capita GSDP (Rs.)")
plt.xlabel("State")
plt.title("Top 5 States based on Total GDP")
plt.show()
```

```python
#bottom 5
df_4=df_3.sort_values("Per Capita GSDP (Rs.)", ascending=True)
new_df_1=df_4.head()

plt.figure(figsize=(9,4))
sns.set_style("whitegrid")
sns.barplot(x=new_df_1.index,y=new_df_1["Per Capita GSDP (Rs.)"],data=new_df_1)
plt.ylabel("Per Capita GSDP (Rs.)")
plt.xlabel("State")
plt.title("Bottom 5 States based on Total GDP")
plt.show()
```

```python
#ratio of the highest per capita GDP to the lowest per capita
df_4=df_3.sort_values("Per Capita GSDP (Rs.)", ascending=False)
df_5=df_4.max()/df_4.min()
df_5["Per Capita GSDP (Rs.)"]
```

```python

s=pivot.iloc[[9,15,17,24,31],:]

sf=s.T
# Create Primary percentage

sf["Primary %"] = sf['Primary']/sf['Gross State Domestic Product']*100
sf.head()
```

```python
#create Secondary percentage
sf["Secondary%"] = sf['Secondary']/sf['Gross State Domestic Product']*100
sf.head()
```

```python
#create  Tertiary percentage
sf["Tertiary%"] = sf['Tertiary']/sf['Gross State Domestic Product']*100
sf.head()
```

```python
pf= sf
#Drop all  the colums "Gross State Domestic Product","Per Capita GSDP (Rs.)","Primary", "Secondary", "Tertiary"
sf_1= pf.drop(columns=["Gross State Domestic Product", "Per Capita GSDP (Rs.)","Primary", "Secondary", "Tertiary"])
#Print Remaining % columns
sf_1.head(20)
```

```python
# Plotting the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.
plt.figure(figsize=(12,12))
plt.title("Percentage Contribution of Primary, Secondary and Tertiary", size = 23)
sf2 = sns.heatmap(sf_1,cmap="viridis", annot = True )
```
